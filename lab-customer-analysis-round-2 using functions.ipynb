```python
def fillna(col, method):
    if method == 'mean':
        file[col] = file[col].fillna(np.mean(file[col]))
    elif method== 'med':
        file[col] = file[col].fillna(np.med(file[col]))
    elif method == 'mod':
        file[col] = file[col].fillna(stats.mode(file[col]))    
    else:
        print('Method not defined')
   


```
